#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
# Convertimos la columna 'date_added' a formato datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extraemos el año de la fecha
df['year_added'] = df['date_added'].dt.year

# Agrupamos por año y contamos el número de títulos añadidos
titles_by_year = df.groupby('year_added').size().reset_index(name='count')

# Ordenamos por año de menor a mayor
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?

```python
tv_ma_titles = df[df["rating"] == "TV-MA"]
tv_ma_titles_gb = tv_ma_titles.groupby(["country"])["title"].count().reset_index()
tv_ma_titles_gb = tv_ma_titles_gb.sort_values(by="title", ascending=False)
tv_ma_titles_gb

top_tv_ma_country = tv_ma_titles_gb.iloc[0]
top_tv_ma_country
```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
movies = df[df['type'] == "Movie"]
movies["Runtime"] = movies["duration"].str.extract(r"(\d+)").astype(float)

avg_duration = movies.groupby('country')['duration'].mean().reset_index
avg_duration_by_country = avg_duration.sort_values(by="duration", ascending=False)
```
