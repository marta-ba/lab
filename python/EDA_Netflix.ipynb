# Uni√≥n de base de datos

- Se hace un LEFT JOIN para ver qu√© titulos en el cat√°logo general (netflix_titles) no son originales de Nertflix. 

```python
netflix_originals.rename(columns={'Title': 'title'}, inplace=True)
merged_data = pd.merge(netflix_titles, netflix_originals, how='left', on='title')
```
