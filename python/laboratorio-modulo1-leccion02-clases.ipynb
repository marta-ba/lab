````python
class Cancion:
    def __init__(self, titulo, artista, album, genero, duracion):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.genero = genero
        self.duracion = duracion

    def mostrar_informacion(self):
        print(f"Título: {self.titulo}, Artista: {self.artista}, Álbum: {self.album}, Género: {self.genero}, Duración: {self.duracion} min")

    def comparar_duracion(self, otra_cancion):
        if self.duracion > otra_cancion.duracion:
            print(f"La canción '{self.titulo}' es más larga que '{otra_cancion.titulo}'")
        elif self.duracion < otra_cancion.duracion:
            print(f"La canción '{self.titulo}' es más corta que '{otra_cancion.titulo}'")
        else:
            print(f"La canción '{self.titulo}' y '{otra_cancion.titulo}' tienen la misma duración")
        
    def calcular_popularidad(self):
        popularidad = 50
        if self.duracion <= 4:
            popularidad += 20
        else:
            popularidad -= 10
        
        if self.genero in ["Pop", "Rock", "Hip-Hop", "Electrónica"]:
            popularidad += 20
        else:
            popularidad -= 5
        
        if self.album != "":
            popularidad += 10
        
        print (f" la canción '{self.titulo}' tiene una popularidad de {popularidad} puntos")

cancion_led_zeppelin = Cancion("Stairway to Heaven", "Led Zeppelin", "Led Zeppelin IV", "Rock", 8.02)
cancion_marvin_gaye = Cancion("What's Going On", "Marvin Gaye", "What's Going On", "Soul", 3.54)

cancion_led_zeppelin.calcular_popularidad()
cancion_marvin_gaye.calcular_popularidad()
````

## Ejercicio 2. Definición de la subclase Colaboracion
2.1. Define  la subclase Colaboracion, la cual va a tener 5 atributos heredados de la clase Cancion (titulo, artista, album, duracion y genero) y va a tener un atributo básico (artistas_colaboradores)


2.2. Crea un método que nos permita mostrar todos los detalles de una canción concreta (incluidos los artistas colaboradores).

2.3. Crea un método que nos permita comprobar si una canción es o no una colaboración

2.4. Crea un método que nos permita obtener el número de colaboradores.

2.5. Crea un método que nos actualice la puntuación de la canción, ya qu es una colaboración sumará 5 puntos por cada colaborador

2.6. Crea un método que nos permita obtener el colaborador principal asumiendo que es el primer colaborador en la lista.

2.7. Crea una instancias de la clase Colaboracion con la siguiente info: ("Meant to Be", "Bebe Rexha", "Expectations", 2.44, "Country Pop", ["Florida Georgia Line"])

2.8. Comprueba el funcionamiento de cada uno de los métodos de la subclase


````python
class Colaboracion(Cancion):
    def __init__(self, titulo, artista, album, duracion, genero, artistas_colaboradores):
        super().__init__(titulo, artista, album, duracion, genero)
        self.artistas_colaboradores = artistas_colaboradores

    def mostrar_detalles(self):
        print(f"Título: {self.titulo}")
        print(f"Artista: {self.artista}")
        print(f"Álbum: {self.album}")
        print(f"Género: {self.genero}")
        print(f"Duración: {self.duracion} minutos")
        print(f"Colaboradores: {', '.join(self.artistas_colaboradores)}")
    
    def es_colaboracion(self):
        if len(self.artistas_colaboradores) > 0:
            print("Esta canción es una colaboración.")
        else:
            print("Esta canción no es una colaboración.")
    
    def obtener_numero_colaboradores(self):
        return len(self.artistas_colaboradores)
    
    
    def actualizar_puntaje(self):
        puntuacion_base = 50
        puntuacion_total = puntuacion_base + (5 * len(self.artistas_colaboradores))
        return puntuacion_total
   
    def obtener_colaborador_principal(self):
        if self.artistas_colaboradores:
            return self.artistas_colaboradores[0]
        else:
            return "No hay colaboradores."
        
    
cancion_colaboracion = Colaboracion("Meant to Be", "Bebe Rexha", "Expectations", 2.44, "Country Pop", ["Florida Georgia Line"])
cancion_colaboracion.mostrar_detalles()
cancion_colaboracion.es_colaboracion()



````

````python
class Cancion:
    def __init__(self, titulo, artista, album, duracion, genero):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.duracion = (duracion)
        self.genero = genero

    def mostrar_informacion(self):
        print(f"Título: {self.titulo}, Artista: {self.artista}, Álbum: {self.album}, Género: {self.genero}, Duración: {self.duracion} min")



class CancionPremiada(Cancion): #creamos una subclase heredada de cancion 
    def __init__(self, titulo, artista, album, duracion, genero, premios):
        super().__init__(titulo, artista, album, duracion, genero) #usamos super para llamar al constructor
        self.premios = premios  #atributo nuevo


    def mostrar_detalles(self): #metodo q me muestre los detalles de la cancion 
        self.mostrar_informacion()
        print(f"Premios ganados: {', '.join(self.premios)}")  # Mostrar los premios


    def obtener_numero_premios(self): #metodo q me muestre el numero de premios
        return len(self.premios)  #devolvemos la longitud de la lsita premios

    def tiene_premio_prestigioso(self): #metodo para saber q premios tiene
        lista_premios = ["Grammy", "Oscar", "Golden Globe", "Brit Award"]
        premios_prestigiosos = [premio for premio in self.premios if premio in lista_premios] #tiene premios prestigiosos?
        if premios_prestigiosos:
            print(f"La canción ha ganado los siguientes premios prestigiosos: {', '.join(premios_prestigiosos)}")
        else:
            print("La canción no ha ganado premios prestigiosos.")


cancion_premiada = CancionPremiada("Just the Way You Are", "Bruno Mars", "Doo-Wops & Hooligans", 3.40, "Pop", ["Grammy", "American Music Award"])


cancion_premiada.mostrar_detalles()
print(f"Número de premios: {cancion_premiada.obtener_numero_premios()}")
cancion_premiada.tiene_premio_prestigioso()
````

````python
class Cancion:
    def __init__(self, titulo, artista, album, genero, duracion):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.genero = genero
        self.duracion = duracion

class Playlist:
    def __init__(self, nombre):
        self.nombre = nombre  
        self.canciones = [] #lista vacía que se llenará con objetos de la clase canción

    def agregar_cancion(self, cancion):
        if isinstance(cancion, Cancion): 
            self.canciones.append(cancion)  
            print(f"La canción '{cancion.titulo}' ha sido añadida a la playlist '{self.nombre}'.")
        else:
            print("El objeto no es una canción válida.")

    def mostrar_playlist(self):
        if self.canciones:
            print(f"Playlist: {self.nombre}")
            for cancion in self.canciones:
                cancion.mostrar_informacion()
        else:
            print(f"La playlist '{self.nombre}' está vacía.")
    
    def ver_playlist(self):
        print(f"Playlist: {self.nombre}")
        for cancion in self.canciones:
            if cancion.artista == artista:
                print(cancion)
    
    def contar_canciones_y_duracion(self):
        total_duracion = sum([c.duracion for c in self.canciones])
        print(f"Número total de canciones: {len(self.canciones)}")
        print(f"Duración total: {total_duracion} min")

    def ordenar_aleator(self):
        random.shuffle(self.canciones)
````
